```python
# code here

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)
cos = np.cos(x)
cos3 = np.cos(x)*3
sin = np.sin(x-90)
sin2 = np.sin(x)*2

plt.plot(x, cos, label='Cosx', color='red')
plt.plot(x, cos3, label='3Cosx', linestyle='dashed', color='green')
plt.plot(x, sin, label='Sin(x-90)', linestyle='dashed', color='blue')

list_marker = ['x']
for marker in list_marker:
    plt.plot(x, sin2, marker=marker, linestyle='None', label='2Sinx', color='yellow')

plt.legend(loc='center')
plt.show()
```
