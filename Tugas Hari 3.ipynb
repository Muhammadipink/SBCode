<h2>Soal 1: Comparison Operator</h2>

- Berikan contoh comparison dari tipe data string
- Berikan contoh comparison dari tipe data boolean
- Berikan contoh comparison dari tipe data integer

```python
# data string 

x = 'mobil'
if x == 'mobil': #operasi == atau sama dengan
    print('A. comparison data string.','\n', 
    "x = ",x,type(x))

# data boolean
mobil = True
motor = False
kendaraan = mobil != motor #operasi != atau tidak sama dengan
print('B. comparison data boolean.','\n',
'Apakah mobil tidak sama dengan motor =>',kendaraan, type(kendaraan))

# data integer
hari = 31
bulan = 12
hasil = bulan > hari #operasi > atau lebih besar dari
print('C. comparison data integer.','\n',
'Apakah jumlah bulan lebih besar dari jumlah hari: ', hasil, type(hari))
```

<h2>Soal 5: While Loop dan For Loop</h2>

- Apa perbedaan while loop dan for loop?
- Berikan contoh sederhana cara menggunakan while loop dan for loop

> 1. **While Loop** adalah jenis pengulangan yang jumlah pengulangannya tak terbatas selama masih memenuhi kondisi yang telah ditentukan sebelumnya.
> 2. **For Loop** adalah jenis pengulangan yang jumlah pengulangannya sebanyak jumlah elemen dalam variabel yang ingin diproses

```python
# contoh while loop
i = 1
while i <= 3:
    print("Ini contoh while loop")
    i += 1

# contoh for loop
hewan = ['Kucing','Anjing','Kelinci','Ayam']
for i in hewan:
    print(i)
```
